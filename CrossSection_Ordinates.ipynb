```python
import math
import csv

# Function to calculate new coordinates based on distance and bearing
def calculate_new_coordinates(lat, lon, distance, bearing):
    R = 6371.0  # Radius of the Earth in kilometers

    # Convert latitude and longitude from degrees to radians
    lat_rad = math.radians(lat)
    lon_rad = math.radians(lon)
    bearing_rad = math.radians(bearing)

    # Calculate the new latitude
    new_lat_rad = math.asin(math.sin(lat_rad) * math.cos(distance / R) +
                            math.cos(lat_rad) * math.sin(distance / R) * math.cos(bearing_rad))

    # Calculate the new longitude
    new_lon_rad = lon_rad + math.atan2(math.sin(bearing_rad) * math.sin(distance / R) * math.cos(lat_rad),
                                       math.cos(distance / R) - math.sin(lat_rad) * math.sin(new_lat_rad))

    # Convert the new coordinates from radians to degrees
    new_lat = math.degrees(new_lat_rad)
    new_lon = math.degrees(new_lon_rad)

    return new_lat, new_lon

# Data provided by the user
distances = [
    0, 20, 40, 70, 100, 140, 180, 220, 250, 290, 330, 370, 400, 440, 470, 520,
    560, 565, 600, 640, 680, 720, 760, 790, 804, 808, 812, 817, 850, 890, 930,
    950, 980, 990, 996, 1000, 1005, 1010, 1040, 1060, 1080, 1100, 1120, 1140,
    1155, 1170, 1180, 1190, 1195, 1200, 1206, 1215, 1230, 1260, 1300, 1340,
    1380, 1420, 1450, 1490, 1530, 1560, 1600, 1640, 1680, 1720, 1750, 1780,
    1810, 1830, 1850, 1870, 1882, 1885, 1890, 1900, 1910, 1920, 1935, 1950,
    1965, 1980, 1990, 2000, 2010, 2020, 2030, 2040, 2050, 2060, 2065, 2068,
    2072, 2085, 2100
]




# Initial coordinates and bearing angle
initial_lat = 25.929055
initial_lon = 89.535027
bearing_angle = 251

# Calculate new coordinates for each distance
new_coordinates = []
for distance in distances:
    distance_km = distance / 1000.0  # Convert meters to kilometers
    new_lat, new_lon = calculate_new_coordinates(initial_lat, initial_lon, distance_km, bearing_angle)
    new_coordinates.append((distance, new_lat, new_lon))

# Write the corrected coordinates to a CSV file
with open('corrected_coordinates10.csv', 'w', newline='') as csvfile:
    fieldnames = ['Distance (m)', 'Latitude', 'Longitude']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for distance, new_lat, new_lon in new_coordinates:
        writer.writerow({'Distance (m)': distance, 'Latitude': new_lat, 'Longitude': new_lon})

print("The corrected coordinates have been written to 'corrected_coordinates10.csv'.")
```
